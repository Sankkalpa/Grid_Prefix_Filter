### Read the number of Seen Ids from the Program

```python
#To count the number of queries in the dataset and initialize the seen_id_per_query array to 0 for each query
seen_id_per_query = []

file_path = '/home/sp5rt/LSH_SIM_CEM/similarity_search_project/geometric-ANN-main/src/data/seen_id_lengths/seen_id_lengths0.csv'
with open(file_path, mode='r') as file:
    csvreader = csv.reader(file)  # Create a CSV reader object

    # Read first row and store it
    for row in csvreader:
            # Optionally, print each row to see the data
        for query in row:
            seen_id_per_query.append(0)

print(len(seen_id_per_query))

```

```python
#Filling the seen_id_per_query array with the number of seen ids for each query

for i in range(100):
    file_path = f'/home/sp5rt/LSH_SIM_CEM/similarity_search_project/geometric-ANN-main/src/data/seen_id_lengths/seen_id_lengths{i}.csv'
    with open(file_path, mode='r') as file:
        csvreader = csv.reader(file)  # Create a CSV reader object
        for row in csvreader:
            # Optionally, print each row to see the data
            for i in range(len(row)):
                seen_id_per_query[i] += int(row[i])

print(seen_id_per_query)
```
